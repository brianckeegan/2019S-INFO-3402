```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb

import numpy as np
import pandas as pd

pd.options.display.max_columns = 200

import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# Read in data
df = pd.read_excel('state_govt_tax_collections.xls',sheet_name=0,header=[1],thousands=',',na_values='X')

# Rename columns
df.rename(columns={'Unnamed: 0':'Year','Unnamed: 1':'Index','Unnamed: 2':'State','Unnamed: 3':'FY'},inplace=True)

# Drop useless columns
df.drop(['Index','FY'],axis=1,inplace=True)

# Extract the state name
df['State'] = df['State'].str.extract(r'([A-Z\s]{2})')

# Drop columns that have US totals and happened before 1950
df = df[(df['State'] != "US") & (df['Year'] > 1950)]

# Inspect
df.head()
```
