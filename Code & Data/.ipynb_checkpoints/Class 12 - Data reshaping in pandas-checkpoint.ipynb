[Tuberculosis](https://en.wikipedia.org/wiki/Tuberculosis) (TB) [killed 1.6 million people](https://www.who.int/en/news-room/fact-sheets/detail/tuberculosis)—including 230,000 children—in 2017 alone. The World Health Organization collects up-to-date information and publishes historical data about the prevalence of TB by country. I have taken their [raw data on case notifications](https://extranet.who.int/tme/generateCSV.asp?ds=notifications) and cleaned it up (see Appendix below) into tidier data for us experiment with different data manipulation functions in `pandas`.

```python
tb_df = pd.read_csv('who_tuberculosis_cleaned.csv')
tb_df.tail()
```

The case data we're interested in was only really available between 2006 and 2012. Let's filter our data down to only these years, but there should still be a fair amount of missing data.

```python
tb_f_df = tb_df.query('2005 < year < 2013')
tb_f_df.tail()
```

## Visualizing tidy data with `seaborn`

With the data in this tidy format, we can perform some more complex visualizations using `seaborn`'s [`catplot`](https://seaborn.pydata.org/tutorial/categorical.html).

```python
g = sb.catplot(x='year',y='cases',hue='age',col='sex',data=tb_f_df,kind='bar',aspect=2)

# Re-scale axes to log-scale to see differences
axs = g.axes.flat
for ax in axs:
    ax.set_ylim((1e1,1e4))
    ax.set_yscale('log')
```

We can also try another orientation, hue, and kind of plot.

```python
g = sb.catplot(x='year',y='cases',hue='region',row='sex',data=tb_f_df,kind='point',aspect=2,dodge=True)

# Re-scale axes to log-scale to see differences
axs = g.axes.flat
for ax in axs:
    ax.set_ylim((1e1,1e4))
    ax.set_yscale('log')
```

## Pivoting tidy data
This "tidy" data has many of the benefits outlined in Wickham's paper, but is difficult to summarize. Let's use a pivot table to track the incidence by country and year.

```python
tb_df_pivoted = pd.pivot_table(tb_f_df,index='year',columns='country',values='cases',aggfunc='sum')
tb_df_pivoted.head(10)
```

```python
ax = tb_df_pivoted['Singapore'].plot(legend=False)

ax.set_yscale('log')
```

We can access a single row of data with `.loc`.

```python
tb_df_pivoted.loc[2012]
```

We can also access a single column.

```python
tb_df_pivoted.loc[:,'United States of America']
```

```python
tb_df_pivoted['United States of America']
```

We also could have pivoted by region.

```python
tb_df_pivoted = pd.pivot_table(tb_f_df,index='year',columns='region',values='cases',aggfunc='sum')
tb_df_pivoted
```

```python
tb_df_pivoted.plot()
```

## Stacking and unstacking data

`multiindex_tb_df` is an example of a pandas Series with a [MultiIndex](https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html). This MultiIndex has 4 levels (sort of like columns): country, year, sex, and age.

![Unstacking data in pandas](http://pandas.pydata.org/pandas-docs/stable/_images/reshaping_unstack.png)

```python
mi_tb_df = tb_f_df.set_index(['country','year','sex','age'])['cases'].sort_index()
mi_tb_df.head(30)
```

We can use `.unstack()` to turn one of the levels inside the MultiIndex into a column. Let's unstack the "sex" level.

```python
mi_tb_df.unstack('sex')
```

We could unstack multiple levels to create a multi-index along the columns as well.

```python
mi_tb_df_sex_age = mi_tb_df.unstack(['sex','age'])
mi_tb_df_sex_age.head(20)
```

We can access the rows in this complex multi-index DataFrame by passing a tuple.

```python
mi_tb_df_sex_age.loc[('Albania',2010)]
```

We can access the columns also with a tuple.

```python
mi_tb_df_sex_age.loc[:,('Female','15-24')]
```

```python
mi_tb_df_sex_age[('Female','15-24')]
```

If we wanted to turn this back into a slightly simpler DataFrame, we could stack the data.

![Stacking data in pandas](http://pandas.pydata.org/pandas-docs/stable/_images/reshaping_stack.png)

Let's stack the sex back into the index and keep the ages as columns.

```python
mi_tb_df_sex_age.stack('sex').head(20)
```

## Melting data

![Melting data in pandas](http://pandas.pydata.org/pandas-docs/stable/_images/reshaping_melt.png)

Read in the data from Wickham's GitHub repo.

```python
billboard_raw_df = pd.read_csv('https://github.com/hadley/tidy-data/raw/master/data/billboard.csv',encoding='latin1')
billboard_raw_df.head()
```

We want to melt all those "x1st.week" through "x76th.week" columns into a tidier format.

```python
id_vars = billboard_raw_df.columns[:7]

billboard_melted_df = pd.melt(billboard_raw_df,id_vars=id_vars)

billboard_melted_df.head(20)
```

We can do a bit more cleanup by changing the names of the columns and extracting more specific information.

```python
billboard_melted_df.rename(columns={'variable':'week','value':'rank'},inplace=True)
billboard_melted_df.head(20)
```

Get the week number out with a regular expression for `pandas`'s [`extract`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.extract.html) method.

```python
billboard_melted_df['week'] = billboard_melted_df['week'].str.extract(r'x(\d+)')
billboard_melted_df.head(20)
```

Some of these values are strings rather than numbers, so cast them to numbers so we can plot.

```python
# Turn the track lengths into seconds (Python interprets the time as hours and minutes initially...)
billboard_melted_df['time'] = pd.to_datetime(billboard_melted_df['time']).apply(lambda x:x.hour*60 + x.minute)

# Turn the week into an int
billboard_melted_df['week'] = billboard_melted_df['week'].apply(int)

billboard_melted_df.head(20)
```

Now it's easy to filter what are the top albums and when. It only took Madonna's "[Music](https://www.youtube.com/watch?v=Sdz2oW0NMFk)" 6 weeks to reach #1 while it took Lonestar's "[Amazed](https://www.youtube.com/watch?v=x-skFgrV59A)" 40 weeks to reach #1.

```python
billboard_melted_df.query('rank == 1')
```

Alternatively, you could look at the trajectory of a single song.

```python
ax = billboard_melted_df.query('track == "Music"').plot(x='week',y='rank',lw=3)
ax.set_xlim((0,30))
ax.set_ylim((50,0))
```

Use `seaborn` again to plot out some relationships.

```python
# Make the catplot
g = sb.catplot(x='week',y='rank',hue='genre',data=billboard_melted_df,aspect=3,kind='point')

# Pull the axis out of the FacetGrid returned by catplot
ax = g.axes.flat

# Reverse the y-axis
ax[0].set_ylim((100,0))
```

Pivot the data so that the tracks are columns, their position by week are indexes, and the values are their rank that week.

```python
billboard_track_week_pivoted = billboard_melted_df.pivot_table(index='week',columns='track',values='rank')
billboard_track_week_pivoted.head(20)
```

Count up the number of weeks a track was in the Billboard Top 100. Only save the ones that spend more than 26 weeks in the Top 100.

```python
top_tracks = billboard_track_week_pivoted.notnull().sum()
top_tracks = top_tracks[top_tracks > 26]
top_tracks = top_tracks.index
top_tracks
```

Plot out these top tracks.

```python
f,ax = plt.subplots(1,1,figsize=(12,4))
ax = billboard_track_week_pivoted[top_tracks].plot(legend=False,ax=ax,lw=3,cmap='rainbow')
ax.set_ylim((100,0))
ax.set_ylabel('Rank')
ax.set_title('Billboard Hot 100 ranks by week since release in 2000')
```

## Appendix

This data comes from the [World Health Organization's tuberculosis project](https://www.who.int/tb/country/data/download/en/).

```python
raw_df = pd.read_csv('https://extranet.who.int/tme/generateCSV.asp?ds=notifications')

raw_df.tail()
```

Extract the relevant columns about new pulmonary smear positive cases as well as other information about countries, regions, and years.

```python
age_cols = ['new_sp_m04','new_sp_m514', 'new_sp_m1524', 'new_sp_m2534', 'new_sp_m3544', 'new_sp_m4554',
            'new_sp_m5564', 'new_sp_m65', 'new_sp_f04', 'new_sp_f1524', 'new_sp_f2534', 'new_sp_f3544',
            'new_sp_f4554', 'new_sp_f5564', 'new_sp_f65']

other_cols = ['country','g_whoregion','year']

all_cols = other_cols + age_cols

ep_df = raw_df[all_cols]
ep_df.tail()
```

Melt the data down and rename the columns.

```python
ep_df_melted = pd.melt(ep_df,id_vars=other_cols,value_vars=age_cols)

ep_df_melted.columns = ['country','region','year','sex_age','cases']

ep_df_melted.head()
```

Extract the sex and the age from the "sex_age" column, drop the "sex_age" column, and save to disk.

```python
ep_df_melted['sex'] = ep_df_melted['sex_age'].str.extract(r'new_ep_([mf])\d+')
ep_df_melted['sex'] = ep_df_melted['sex'].replace({'m':'Male','f':'Female'})

ep_df_melted['age'] = ep_df_melted['sex_age'].str.extract(r'new_ep_[mf](\d+)')
ep_df_melted['age'] = ep_df_melted['age'].replace({'04':'0-4','514':'5-14','1524':'15-24','2534':'25-34',
                                                   '3544':'35-44','4554':'45-54','5564':'55-64','65':'65+'})

del ep_df_melted['sex_age']

ep_df_melted.to_csv('who_tuberculosis_cleaned.csv',index=False)

ep_df_melted.head()
```
