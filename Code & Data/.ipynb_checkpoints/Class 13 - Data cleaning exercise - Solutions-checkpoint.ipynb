## Fix the columns and index

(Hint: The easiest way is to read in the data again, but use different parameters. Read the documentation!)

```python
passengers_df = pd.read_excel('DIA_activity.xlsx',sheet_name='Passengers',
                             header=[0,1],index_col=0)

passengers_df.head()
```

## Melt the data down

```python
passengers_melted = passengers_df.stack([0,1])
passengers_melted.head(20)
```

## Make sure each variable has its own column... and rename the columns

```python
passengers_melted2 = passengers_melted.reset_index()
passengers_melted2.columns = ['date','parent_carrier','carrier','passengers']
passengers_melted2.head(20)
```

## Remove the "Unnamed: N_level_1" rows

```python
passengers_melted3 = passengers_melted2[~passengers_melted2['carrier'].str.contains('Unnamed: ')]
passengers_melted3.head(20)
```

## Separate the "date" column into year and months

(Normie: Access the month and year as attributes of a datetime/timestamp)  
(Elite: Use `.str.extract` and pass a regular expression that matches a 4 digit number for year and 2 digit number of month)

```python
passengers_melted3['year'] = passengers_melted3['date'].str.extract(r'(\d{4})')
passengers_melted3['month'] = passengers_melted3['date'].str.extract(r'-(\d{2})')
passengers_melted3['date'] = pd.to_datetime(passengers_melted3['date'])
```

```python
passengers_melted3.head()
```

## Pivot the data into total passengers by parent carrier per year

```python
passengers_pivot_df = pd.pivot_table(passengers_melted3,
                                     values='passengers',
                                     index='year',
                                     columns='parent_carrier',
                                     aggfunc='sum')

passengers_pivot_df
```

## Plot the pivot table

Is the drop-off in 2018 "real"?

```python
f,ax = plt.subplots(1,1,figsize=(12,4))
ax = passengers_pivot_df.plot(lw=3,legend=False,ax=ax)
ax.legend(loc='center left',bbox_to_anchor=(1,.5))
```

## Plot data

Make a `catplot` with "month" on the x-axis, "passengers" on the y-axis, and hues for the different parent carriers (probably excluding "Other"). What is the top month (on average) for passengers into DIA?

```python
g = sb.catplot(x='month',y='passengers',hue='parent_carrier',kind='point',
               data=passengers_melted3.query('parent_carrier != "Other"'),
               aspect=2,dodge=True)

ax = g.axes[0,0]

# ax.set_ylim((1e3,1e7))
# ax.set_yscale('log')
```
